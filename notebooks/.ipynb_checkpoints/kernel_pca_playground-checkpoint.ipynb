```python

np.random.seed(0)

X, y = make_circles(n_samples=400, factor=.3, noise=.05)

reds = y == 0

blues = y == 1
```

```python
class myKernelPCA(KernelPCA):
    def __init__(self, n_components=None, kernel="linear",
                 gamma=None, degree=3, coef0=1, kernel_params=None,
                 alpha=1.0, fit_inverse_transform=False, eigen_solver='auto',
                 tol=0, max_iter=None, remove_zero_eig=False,
                 random_state=None, copy_X=True, n_jobs=None, k=1):
        self.k = k
        super(myKernelPCA, self).__init__(n_components=None,  kernel="linear",
                 gamma=None, degree=3, coef0=1, kernel_params=None,
                 alpha=1.0, fit_inverse_transform=False, eigen_solver='auto',
                 tol=0, max_iter=None, remove_zero_eig=False,
                 random_state=None, copy_X=True, n_jobs=None)
        
    def fit(self, X, y=None):
        super(myKernelPCA, self).fit(X, y)
        self.alphas_null = self.alphas_[:,self.k:]
        self.lambdas_null = self.lambdas_[self.k:]
        
        self.alphas_ = self.alphas_[:,:self.k]
        self.lambdas_ = self.lambdas_[:self.k]
        return self
    
    def to_nullspace(self, X):
        K = self._get_kernel(X, self.X_fit_)
        K = self._centerer.transform(K)
        
        
        non_zeros = np.flatnonzero(self.lambdas_null)
        scaled_alphas = np.zeros_like(self.alphas_null)
        scaled_alphas[:, non_zeros] = (self.alphas_null[:, non_zeros]
                                       / np.sqrt(self.lambdas_null[non_zeros]))
        
        return np.dot(K, scaled_alphas)

    def fit_my_inverse_transform(self, X):
        """
        Ideally can be fitted on set larger than the original 10 point
        training set from Bolukbasi et al. Does not project X into subspace
        to learn the kernel directly instead learns (x, phi(x)) pairs.
        """
        if hasattr(X, "tocsr"):
            raise NotImplementedError("Inverse transform not implemented for "
                                      "sparse matrices!")

        n_samples = X.shape[0]
        null_transformed = self.to_nullspace(X)
        X_transformed = self.transform(X)
        K1 = 0.95 * self._get_kernel(X_transformed)
        K2 = 0.05 * self._get_kernel(null_transformed)
        K = np.concatenate((K1, K2), axis=1)
        print(K.shape, "kshape")
        Kpsd = K.T.dot(K)
        print(Kpsd.shape, "kshape2")

        Kpsd = Kpsd + self.alpha * np.eye(Kpsd.shape[0])
        
        self.dual_coef_ = linalg.solve(Kpsd, K.T.dot(X), sym_pos=True, overwrite_a=True)
        print(self.dual_coef_.shape, "dualcoef")
        self.dual_coef_range = self.dual_coef_[:n_samples, :]
        self.dual_coef_nullspace = self.dual_coef_[n_samples:, :]
        print(self.dual_coef_nullspace.shape, "dual_coef_nullspace")
        self.X_transformed_fit_ = X_transformed
        self.X_transformed_fit_null = null_transformed
        self.fit_inverse_transform = True
    
    def inverse_null_transform(self, X):

        if not self.fit_inverse_transform:
            raise NotFittedError("The fit_inverse_transform parameter was not"
                                 " set to True when instantiating and hence "
                                 "the inverse transform is not available.")
        
        null = 0.05 * self._get_kernel(self.to_nullspace(X) , self.X_transformed_fit_null)
        range_ = 0.95 * self._get_kernel(self.transform(X), self.X_transformed_fit_)
        conc =  np.concatenate((range_, null), axis=1)
        K = conc
#         return np.dot(K, self.dual_coef_)
        comp1 = np.dot(null, self.dual_coef_nullspace)
        comp2 =  np.dot(range_, self.dual_coef_range)
#         return comp1 + comp2, self.to_nullspace(X), self.transform(X)
        return comp1 + comp2, comp1, comp2
```
